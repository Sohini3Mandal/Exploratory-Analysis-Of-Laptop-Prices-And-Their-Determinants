# **EXPLORATORY DATA ANALYSIS ON LAPTOP PRICING DATA**

**Dataset Name**

Laptop Pricing.

This dataset is a collection of features related to various laptops.

**About Dataset**

This dataset is a collection of features related to various laptops, such as brand, processor type, RAM, storage capacity, and other specifications. The dataset also includes the corresponding prices of these laptops. This dataset can be used for regression analysis to predict the prices of laptops based on their features. The dataset is suitable for data scientists, machine learning enthusiasts, and researchers who are interested in building regression models to predict the prices of laptops based on various features.

**Parameters**

The parameters used in the dataset are:

1. *Manufacturer.*
The company that manufactured the laptop

2. *Category.*
The category to which the laptop belongs: This parameter is mapped to numerical values in the following way:

Category - Assigned Value

Gaming - 1
Netbook - 2
Notebook - 3
Ultrabook - 4
Workstation - 5

3. *GPU.*
The manufacturer of the GPU. This parameter is mapped to numerical values in the following way:

GPU - Assigned Value

AMD - 1
Intel - 2
NVidia - 3

4. *OS.*
The operating system type (Windows or Linux): This parameter is mapped to numerical values in the following way:

OS - Assigned Value

Windows - 1
Linux - 2

5. *CPU_core.*
The type of processor used in the laptop: This parameter is mapped to numerical values in the following way:

CPU_core - Assigned Value

Intel Pentium i3 - 3
Intel Pentium i5 - 5
Intel Pentium i7 - 7

6. *Screen_Size_cm.*
The size of the laptop screen is recorded in cm.

7. *CPU_frequency.*
The frequency at which the CPU operates, in GHz.

8. *RAM_GB.*
The size of the RAM of the system in GB.

9. *Storage_GB_SSD.*
The size of the SSD storage in GB is installed in the laptop.

10. *Weight_kg.*
The weight of the laptop is in kgs.

11. *Price.*
The price of the laptop is in USD.

12. *Screen.*
Categories: IPS Panel,Full HD

```python
cov_matrix=data.cov(numeric_only=True)
cov_matrix
```

The covariance matrix shows that the variances of the features differ substantially, indicating that the variables are measured on different scales. As a result, covariance values are not directly comparable across variable pairs.

While the covariance matrix provides information on the direction and magnitude of joint variability between variables, its values are scale-dependent and not directly comparable. The correlation matrix standardizes these relationships, allowing meaningful comparison of the strength of linear associations across variables, making it more suitable for interpretation and feature selection.

```python
corr_matrix = data.corr(numeric_only=True)
corr_matrix
```

```python
plt.figure(figsize=(20, 10))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Plot",fontsize=25)
```

**Correlation Analysis of Numerical Features**

The correlation heatmap summarizes the pairwise linear relationships among the numerical variables: Screen_Size_cm, CPU_frequency, RAM_GB, Weight_kg, and Price.

Price shows its strongest positive association with RAM_GB (r ≈ 0.57), indicating that laptops with higher memory capacity tend to be priced higher. A moderate positive correlation is also observed between Price and CPU_frequency (r ≈ 0.39), suggesting that processor specifications contribute meaningfully to pricing, though they do not fully explain it. In contrast, Price and Weight_kg exhibit a near-zero correlation (r ≈ −0.04), implying that laptop weight has little linear influence on price.

Among the hardware features, Screen_Size_cm and Weight_kg show a strong positive correlation (r ≈ 0.83), reflecting the intuitive relationship that larger screens are associated with heavier devices. Screen size has negligible correlation with most other variables, including price, indicating limited direct impact on pricing.

Overall, the heatmap indicates that memory capacity and processor characteristics are the primary drivers of laptop price, while physical attributes such as weight and screen size play a comparatively minor role. The absence of extremely high correlations among most predictors suggests limited multicollinearity, supporting their joint use in regression modeling.
